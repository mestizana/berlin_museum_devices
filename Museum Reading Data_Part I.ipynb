```python
def convert_to_seconds(time_str):
    minutes_str, seconds_str = time_str.split('m')
    seconds_str = seconds_str.rstrip('s')  

    minutes = int(minutes_str)
    seconds = int(seconds_str)

    total_seconds = (minutes * 60) + seconds

    return total_seconds
```

```python
df1['Time_Column_in_seconds'] = df1['Information Reading'].apply(convert_to_seconds)
```

```python
df1
```

```python
df1['Time_Column_in_seconds_2'] = df1['Device Usage'].apply(convert_to_seconds)
```

```python
df1.drop(columns='Device Usage' , inplace= True)
```

```python
df1.drop(columns='Information Reading', inplace = True)
```
